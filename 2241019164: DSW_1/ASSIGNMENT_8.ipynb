```python
22. Plot the CGPA vs Percentage of Attendance column. Can you conclude any relation between CGPA
vs Percentage of Attendance just by looking at the graph.
```

```python
23. Find the covariance and correlation between CGPA vs Percentage of Attendance. Can you validate
your established relation in previous ques by the covariance and correlation.
```

```python
corr= dt['cgpa'].corr(dt['%of attendence'])
cov=dt['cgpa'].cov(dt['%of attendence'])
print("Covariance between CGPA and Percentage of Attendance:", cov)
print("Correlation between CGPA and Percentage of Attendance:", corr)
```

```python
27. Find the outliers with respect to CGPA column. If any present show their name.
```

```python
Q1 = dt['cgpa'].quantile(0.25)
Q3 = dt['cgpa'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

dt[(dt['cgpa'] < lower_bound) | (dt['cgpa'] > upper_bound)]
```
