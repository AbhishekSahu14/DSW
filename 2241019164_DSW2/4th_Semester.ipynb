1. CDF of Normal
2. area under normal dist. upto x (Normal probability below x)
3. area under normal dist. after x (Normal probability above x)
4. area under normal dist. after x and upto y (Normal probability between x and y)
5. area under normal dist. upto x and after y (Normal probability outside x and y)
6. inverse of cdf of normal
7. find the point x s.t. area is upto x is p(upper bound)
8. find the point x s.t. area is after x is p(lower bound)
9. find the point x and y s.t. area is between x and y is p(two sided bound)
10. find the point x and y s.t. area is before x and after y is p(two sided bound)

```python
import math
import matplotlib.pyplot as plt
def norm_belowcdf(x,m,s):  
     return (1+ math.erf((x-m)/math.sqrt(2)/s))/2
x=[i/10 for i in range(-50,50)]
y1=[norm_belowcdf(i,0,1) for i in x]
plt.plot(x,y1)
plt.show()
z=norm_belowcdf(8,0,1)
z
```

```python
import math
def norm_abovecdf(x,m,s):  
     return 1-(1+ math.erf((x-m)/math.sqrt(2)/s))/2
x=[i/10 for i in range(-50,50)]
y1=[norm_abovecdf(i,0,1) for i in x]
plt.plot(x,y1)
plt.show()
z=norm_abovecdf(8,0,1)
z
```

```python
import math
def norm_betweencdf(x,y,m,s):  
     return norm_belowcdf(x,m,s) - norm_belowcdf(y,m,s)
z=norm_betweencdf(2,1,0,1)
print(z)
```

```python
def norm_outsidecdf(x,y,m,s):  
     return 1- (norm_belowcdf(x,m,s) - norm_belowcdf(y,m,s))
z=norm_outsidecdf(2,1,0,1)
print(z)
```

```python
def norm_cdf(a,m,s):  
     return (1+ math.erf((a-m)/math.sqrt(2)/s))/2
def norm_inversecdf(prob,m=0,s=1,t=0.00001):
    x=-10
    y=10
    while (y-x)>t:
        mid = x+y/2
        p=norm_cdf(mid,0,1)
        if p==prob:
            return mid
        elif p > prob:
            y=mid
        else:
            x=mid
    return y
norm_inversecdf(0.5)
```

```python
def normal_upper(p,m=0,s=1):
    return norm_inversecdf(p)
normal_upper(0.5)
```

```python
def normal_lower(p,m=0,s=1):
    return norm_inversecdf(1-p)
normal_lower(0.158)
```

```python
def normal_twosided(p,m=0,s=1):
    tail_p=(1-p)/2.0
    upper=normal_lower(tail_p,m=0,s=1)
    lower=normal_upper(tail_p,m=0,s=1)
    return lower,upper
normal_twosided(0.95)
```

```python
def normal_othertwosided(p,m=0,s=1):
    tail_p=(1-p)/2.0
    upper=1-normal_lower(tail_p,m=0,s=1)
    lower=1-normal_upper(tail_p,m=0,s=1)
    return lower,upper
normal_othertwosided(0.95)
```

```python
def add_bias(x):
    n=x.shape[0]
    x_b=np.concatenate((np.ones((n,1)),x),axis=1)
    return x_b
```

```python
def cost_logistic(x,y,w,b):
    m=x.shape[0]
    cost = 0
    for i in range(m):
        z_i=np.dot(x[i],w) + b
        f_wb_i=sigmoid(z_i)
        cost += -y[i]*np.log(f_wb_i)-(1-y[i])*np.log(1-f_wb_i)
    cost=cost/m
    return cost
y=np.array([0,0,0,1,1,1])
cost_logistic(x,y,w,b)
```

```python
def gradient_logistic(x,y,w,b):
    m,n=x.shape
    dj_dw=np.zeros((n,))
    dj_db=0.0
    for i in range(m):
        z_i=np.dot(x[i],w) + b
        f_wb_i=sigmoid(z_i)
        err_i=f_wb_i-y[i]
        for j in range(n):
            dj_dw[j] += err_i*x[i,j]
        dj_db += err_i
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    return dj_dw, dj_db
x=np.array([[0.5,1.5],[1,2],[0.1,0.1],[3,0.5],[2,2],[1,2.5]])
y=np.array([0,0,0,1,1,1])
w=np.array([1,1])
b=np.array([-1])
m,n = x.shape
print(m,n)
dj_dw, dj_db=gradient_logistic(x,y,w,b)
print(dj_dw)
print(dj_db)
```

```python
from collections import Counter
def gini_impurity(y):
    counts=Counter(y)
    impurity = 1-sum([(count/len(y))**2 for count in counts.values()])
    return impurity
gini_impurity(['m','m','f','f','f','f'])
```

```python
import math
def entropy(y):
    counts=Counter(y)
    impurity = -sum([(count/len(y))*math.log2(count/len(y)) for count in counts.values()])
    return impurity
entropy(['m','m','m','m','f','f'])
```

```python
import pandas as pd
def compute_gini(data,feature_col,split_value):
    left_branch= data[data[feature_col]<= split_value]
    right_branch= data[data[feature_col]> split_value]
    total=len(data)
    def cal_impurity(branch):
        if len(branch)==0:
            return 0
        counts=Counter(branch['obese'])
        impurity = 1-sum([(count/len(branch))**2 for count in counts.values()])
        return impurity
    gini_left=cal_impurity(left_branch)
    gini_right=cal_impurity(right_branch)
    print(gini_left,gini_right)
    weighted_left=len(left_branch)/total
    weighted_right=len(right_branch)/total
    weighted_avg_gini=(weighted_left*gini_left)+(weighted_right*gini_right)
    return weighted_avg_gini
```
